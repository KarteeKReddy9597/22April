```python
nos = [1,2,3,4,5,6,7,8,9]
```

```python
def odd_nos(nos):
    odd = []
    for no in nos:
        if no%2 == 1:
            odd.append(no)
    return odd
```

```python
odd_nos(nos)
```

```python
def calculate_average_with_for_loop(nos):
    nos_sum = 0
    for no in nos:
        nos_sum = nos_sum + no
    average = nos_sum/len(nos)
    return average
```

```python
%timeit calculate_average_with_for_loop(nos)
```

```python
class Person:
    first_name = ""
    last_name = ""
    age = 0
    
    def __init__(self, first_name, last_name, age):
        self.first_name = first_name
        self.last_name = last_name
        self.age = age
    
    def get_full_name(self):
        return self.first_name + " " + self.last_name
    
class Student(Person):
    roll_no = 0
    fav_subject = ""
    
    def __init__(self, first_name, last_name, age, roll_no, fav_subject):
        super().__init__(first_name, last_name, age)
        self.roll_no = roll_no
        self.fav_subject = fav_subject
    
    def get_roll_no(self):
        return self.roll_no
```

```python
student_obj = Student("Karteek", "Reddy", 21, 12345, "Hindi")
```

```python
student_obj.get_roll_no()
```

```python
student_obj.get_full_name()
```
